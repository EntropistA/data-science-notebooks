# Concepts

ACID - Atomicity, Consistency, Isolation, Durability

Atomicity - operations can be fully complete or not at all

Transactions - set of commands to execute, if one fails - none of them are executed

RDBMS - Relational Database Management System

# Scaling

## Vertically

Improving existing machines (more CPU, RAM)

- expensive

## Horizontally

Adding new machines using a proccess called _Sharding_. Different parts of the data are distributed across many machines.

- cheap

# Database types - SQL, NoSQL

- complementory
- can coexist

## NoSQL

### Types:

#### Key-value

Data is most often stored on server memory, leading to fast operations. However there is a risk of losing data after power shut off. Therefore, most systems provide disk-memory databases. Keys and values can be of any type which can be changed later. 

":" is a common separator in a key, for example: user:{id}:preferences

Usage:
- preferences
- sessions
- shopping carts
- recomendations

Usuited:
- searching values
- relations between databases

Examples:
Redis - Remote Dictionary Server, database, cache, message broker, open-source, Lua scripting, asynchronous replication

Operations:

PUT - inserts new key, value tuple or updates the existing one
GET
DELETE

#### Document

_Documents_ are key-value pairs stored inside a _collection_. Schema does not need to be predefined.


Usage:
- Cataloges
- CMS
- real time analytics

- natural
- human readable format with JSON
- ACID
- replication

Limitations:
- data in application code needs attention to keep it simple and consistent
- with database increasing there could be need to change values in all of the documents, leading to messy code

documents -> rows
collections -> tables

MongoDB offer:
- MQL - Mongo Query Language
- Compass - GUI
- Atlas - Database server for AMS, Azure, Google Cloud
- Atlas Lake - query and analyze data across AWS S3 and Atlas using MQL
- Enterprise Advanced - ability to add MongoDB to the infrastructure
- Charts - creating visualizations
- Realm - store database locally on Android or iOS


#### Column family

Features:
- very fast writing / retrieving
- scales horizontally
- each row may have different number of columns (avoids writing default values)
- each columns has name, value, timestamp (last usage)

Limitations:
- no joins
- no multirow transactions
- no subquery support
- constly design changes

Usage:
- large sets of data
- time series data
- CMS
- event logging

columns family -> table
