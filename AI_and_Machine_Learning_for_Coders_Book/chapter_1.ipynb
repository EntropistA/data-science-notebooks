# Tensor Flow Introduction

## Imports

keras.Sequential - class initiating layers
keras.layers - different layer types

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

# Creating model

nodes -> layer -> Sequential -> model

```python
l0 = Dense(units=1, input_shape=[1])
model = Sequential([l0])
```

Using Stochastic Gradient Descent

INPUTS: real values, previous guess
COMPUTES: loss (error) using "mean_squared_error"
OUTPUTS: new guess

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
xs = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
ys = xs * 2 - 1
print(xs, ys, sep="\n")
```

Fitting a model with input and desired values

```python
model.fit(xs, ys, epochs=500)
```

Predicting a new value

```python
print(model.predict([10.0]))
```

Learning summary

```python
print(f"Here is what I learned: {l0.get_weights()}")
```
